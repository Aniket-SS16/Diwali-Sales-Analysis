```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #Visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# Read the csv file
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
# To avoid encoding error, 'unicode_escape'
```

```python
#To check the number of rows and columns of the imported csv file
df.shape
```

```python
#To display the top 5 rows of the table. You can also enter the value to display as many top rows you want in the brackets like df.head(10) will display top 10 rows
df.head()
```

```python
df.info()
#This will give all the details of the columns. Here we can also find the null columns
```

```python
#Drop unrelated/ blank columns
df.drop(['Status','unnamed1'], axis=1, inplace= True)
#axis = 1 means it will drop the entire row
#inplace= True will delete those columns permanently
```

```python
df.info()
```

```python
#Check for null values in a column
pd.isnull(df).sum()
```

```python
#Drop null values permanently
df.dropna(inplace=True)
```

```python
df.shape
```

```python
#Change the Data Type of a Column
df['Amount'] = df['Amount'].astype('int')
```

```python
#Check the Data Type
df['Amount'].dtypes
```

```python
#List all the columns
df.columns
```

```python
#describe() method returns the description of the data in the DataFrame (i.e. Count, mean, std, etc.)
df.describe()
```

```python
#Use describe() on specific columns
df[['Age','Orders','Amount']].describe()
```

```python
# EXPLORATORY DATA ANALYSIS
```

```python
# Analysis based on GENDER
```

```python
ax = sns.countplot(x= 'Gender', data=df)

#To give the Data Labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Group amount spend by Gender and sort with respect to Amount column
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sns.barplot(x='Gender', y='Amount', data= sales_gen)
```

```python
## From above graphs, we can see that the most of the buyers are Feamles and even the purchasing power of the Females is more than Males
```

```python
# Analysis based on AGE
```

```python
ax = sns.countplot(x= 'Age Group', data=df, hue='Gender')

#To give the Data Labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sns.barplot(x='Age Group', y='Amount', data= sales_age)
```

```python
## From the above graphs, we can see that most of the buyers are from age group 26 to 35 and are females.
```

```python
# Analysis based on STATES
```

```python
# Total Number of Orders from Top 10 States
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
```

```python
az = sns.barplot(x='Orders', y='State', data= sales_state)

#To give the Data Labels
for bars in az.containers:
    az.bar_label(bars)
```

```python
# Total Amount Spent from Top 10 States
sales_amount_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
```

```python
az = sns.barplot(x='Amount', y='State', data= sales_amount_state)
```

```python
## From the above graphs, we can see that most of the Orders and total sales are from Uttar Pradesh, Maharashtra and Karnataka respectively.
```

```python
#Analysis based on Marital Status
```

```python
ax = sns.countplot(x= 'Marital_Status', data=df)
#To adjust the width and height of the bars
sns.set(rc={'figure.figsize':(4,8)})
#To give the Data Labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount Spent from Top 10 States
sales_marital_status = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
```

```python
sns.barplot(data= sales_marital_status, x='Marital_Status', y='Amount', hue='Gender')
```

```python
## From the above graphs, we can see that most of the buyers are married (women) and they have high purchasing power.
```

```python
# Analysis based on Occupation
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x= 'Occupation', data=df)

#To give the Data Labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
az = sns.barplot(x='Occupation', y='Amount', data= sales_occupation)
sns.set(rc={'figure.figsize':(20,5)})
```

```python
## From the above graphs, we can say that most of the buyers are working in IT, Helathcare and Aviation Sectors.
```

```python
#Analysis based on Product Category
```

```python
ax = sns.countplot(x= 'Product_Category', data=df)
sns.set(rc={'figure.figsize':(27,5)})
#To give the Data Labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Top 10 Sales by Product Category
sales_products = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
```

```python
az = sns.barplot(x='Product_Category', y='Amount', data= sales_products)
sns.set(rc={'figure.figsize':(15,5)})
```

```python
## From the above graphs, we can say that most of the sold products are Clothing, Food and Elecronics.
```
